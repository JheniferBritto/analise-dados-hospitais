```python
import pandas as pd

dados = {
    'Hospital ': ['Hospital A', 'hospital a ', 'Hospital B', None, 'Hospital C', 'Hospital D', 'Hospital B'],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Bahia', None, 'Rio de Janeiro', 'Belo Horizonte'],
    'Data Atendimento': ['2024-01-10', '2024/02/15', '15-03-2024', '2024-04-20', '2024-05-01', None, '2024-06-10'],
    'Pacientes Atendidos': ['100', '200', 'erro', '150', '300', '120', None],
    'Custo Total': [30000, 40000, 20000, None, 50000, 25000, 35000]
}

df = pd.DataFrame(dados)
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip().str.title()
```

```python
df['Hospital'] = df['Hospital'].fillna('Desconhecido')
```

```python
df['Cidade'] = df['Cidade'].fillna('Desconhecida')
```

```python
df['Data Atendimento'] = pd.to_datetime(df['Data Atendimento'], errors='coerce')
df['Data Atendimento'] = df['Data Atendimento'].ffill()
```

```python
df['Pacientes Atendidos'] = pd.to_numeric(df['Pacientes Atendidos'], errors='coerce')
df['Pacientes Atendidos'] = df['Pacientes Atendidos'].fillna(df['Pacientes Atendidos'].median())
```

```python
df['Custo Total'] = pd.to_numeric(df['Custo Total'], errors='coerce')
df['Custo Total'] = df['Custo Total'].fillna(df['Custo Total'].median())
```

Qual hospital atende mais pacientes?

```python
df.groupby('Hospital')['Pacientes Atendidos'].sum().sort_values(ascending=False)
```

Qual cidade tem maior custo total?

```python
df.groupby('Cidade')['Custo Total'].sum().sort_values(ascending=False)
```

Qual hospital tem maior custo por paciente?

```python
df.groupby('Hospital')['Custo Total'].sum() / df.groupby('Hospital')['Pacientes Atendidos'].sum().sort_values(ascending=False)
```

Existe algum outlier?

```python
q1 = df['Custo Total'].quantile(0.25)
q3 = df['Custo Total'].quantile(0.75)
iqr = q3 - q1

limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

outlers = df[(df['Custo Total'] < limite_inferior) | (df['Custo Total'] > limite_superior)]
outlers
```
